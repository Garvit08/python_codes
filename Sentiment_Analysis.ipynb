```python
import pandas as pd
import tweepy
from nltk.sentiment.vader  import SentimentIntensityAnalyzer
```

```python
#Twitter API Authenication Variables
consumer_key=''
consumer_secret=''
access_token=''
access_token_secret=''
```

```python
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)

api = tweepy.API(auth)
```

```python
tweets = api.search("Artificial Intelligence" , count = 200)

```

```python
data=[tweet.text for tweet in tweets]
```

```python
data=pd.DataFrame(data)
display(data.head(10))
```

```python

data.columns=["Tweets"]
display(data.head(10))
print(tweets[0].created_at)
```

```python
import nltk
nltk.download("vader_lexicon")
```

```python
sid = SentimentIntensityAnalyzer()

listy = []

for index,row in data.iterrows():
    ss = sid.polarity_scores(row["Tweets"])
    listy.append(ss)
se = pd.Series(listy)
data['polarity']=se.values

display(data.head(100))
```

```python
positive=0
negative=0
neutral=0
compound=0
for a in data['polarity']: 
    positive+=a['pos']
    negative+=a['neg']
    neutral+=a['neu']
    compound+=a['compound']
print("Negative = {}\nPositve = {}\nNeutral = {}\nCompound = {}".format(negative,positive,neutral,compound))
```
